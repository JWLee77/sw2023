####Import sources

```python
import cv2
from google.colab.patches import cv2_imshow
import os
```

####**Image Upload**
#####downloaded from Google Drive

```python
os.chdir("/content/drive/MyDrive/img")
print(os.getcwd())
img=cv2.imread("/content/drive/MyDrive/img/window.jpg")
cv2_imshow(img)
```

####**Convert Color**   
######Gray

```python
gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

####**Blur**

```python
blur = cv2.GaussianBlur(gray, ksize=(3,3), sigmaX=0)
cv2_imshow(blur)
```

####**Canny**

```python
canny=cv2.Canny(blur,50,150)
cv2_imshow(canny)
```
